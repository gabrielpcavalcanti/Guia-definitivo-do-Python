# Tree

A classe Tree da biblioteca Rich permite criar representações visuais de estruturas hierárquicas no terminal, como diretórios, tópicos, categorias ou qualquer dado que tenha relação de pai-filho.

Diferente de listas ou tabelas, Tree mostra profundidade e hierarquia, com linhas de conexão claras e opções de estilo.

É bom usar quando: 

- Visualização clara de hierarquias

- Ideal para diretórios, projetos ou menus

- Permite incluir painéis, tabelas ou textos dentro de cada nó

- Facilita leitura em CLIs complexos

```python
from rich.console import Console
from rich.tree import Tree
from rich.panel import Panel


console = Console()

tree = Tree("Projeto", style="bright_yellow", guide_style="bold bright_blue")


tree.add("src")
tree.add("docs")
tree.add("tests")

src_tree = tree.add("src")
src_tree.add("main.py")
src_tree.add("utils.py")


docs_tree = tree.add("docs")
docs_tree.add("README.md")

painel = Panel(tree, border_style="white", expand=False)


console.print(painel)
```
