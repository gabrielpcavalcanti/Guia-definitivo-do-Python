# Text

Text é um objeto de texto estilizado do Rich.

Tudo que o Rich imprime vira um Text internamente. Text resolve o problema de múltiplos estilos no mesmo texto, sem markup. Fica mais profissional,reutilizavel e maleável.

É preciso importar o módulo para poder utilizar assim como qualquer outro módulo do rich.

Os parâmetros da classe é

```Python
Text(
    text="",
    style="",
    *,
    justify=None,
    overflow=None,
    no_wrap=None,
    end="\n",
    tab_size=None,
    spans=None,
)
```

### 1 - Acrescentar textos com estilos diferentes.

Para isso é preciso criar um objeto do tipo Text e utilizar o método append( ). Assim cada texto pode ter um estilo próprio e sem usar markup.

````python
from rich.console import Console
from rich.text import Text

console = Console()
text = Text()

text.append("Status: ", style="cyan")
text.append("OK", style="bold green")

console.print(text)
````

### 3 - Estilização controlada

Além de inserir texto com a estlização que quiser, existem diversos outros métodos permitem estilizar partes expecíficas do texto, sem alterar o restante.



#### stylize( )

Ele aplica um estilo a um intervalo especificado. Os valores do intervalo são determinados pelo índice da String. 

````python
from rich.console import Console
from rich.text import Text

console = Console()

texto_exemplo: str = "No fim da tarde, a praça parecia suspensa no tempo: crianças corriam atrás de pombos, idosos conversavam nos bancos e o sol tingia tudo de laranja."

texto = Text(texto_exemplo)

texto.stylize("bold green on white", 45, 60)

console.print(texto)
````

#### stylize_before( )

Ele aplica um estilo antes dos estilos que já existem naquele intervalo.

Aplica o estilo antes, com menor prioridade. Se já existir outro estilo depois, o estilo posterior vence.

````python
from rich.text import Text
from rich.console import Console

console = Console()

text = Text("Rich Text")

text.stylize("red", 0, 4)
text.stylize_before("bold green", 0, 4)

console.print(text)
````

#### highlight_words( )

Destaca palavras específicas do texto.

````python
from rich.console import Console
from rich.text import Text

console = Console()

texto_exemplo: str = "No fim da tarde, a praça parecia suspensa no tempo: crianças corriam atrás de pombos, idosos conversavam nos bancos e o sol tingia tudo de laranja."

texto = Text(texto_exemplo)

texto.highlight_words(["tarde", "bancos"], "bold cyan")

console.print(texto)
````

### 4 - Outros métodos

### plain

limpa toda a estilização do texto. É um retorno. Tem que salvar numa variável se quiser, ter o texto sem estilos para depois.

````python
from rich.console import Console
from rich.text import Text

console = Console()

texto_exemplo: str = "No fim da tarde, a praça parecia suspensa no tempo: crianças corriam atrás de pombos, idosos conversavam nos bancos e o sol tingia tudo de laranja."

texto = Text(texto_exemplo)

texto.stylize("red", 0, 16)
texto.stylize("red", 78, 90)

texto.highlight_words(["tarde", "bancos"], "bold cyan")

console.print(texto)

texto_no_stile: str = texto.plain

console.print(texto.plain)

console.print(texto_no_stile)
````
