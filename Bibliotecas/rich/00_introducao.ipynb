# Introdução

A Rich é uma biblioteca Python que transforma seu terminal em algo moderno e visualmente atraente. Ela permite adicionar cores, tabelas, barras de progresso, formatação de texto e muito mais.

O Rich é uma biblioeca que permite fazer muitas coisas, pode desde utilizar os comando mais simples e rápidos até construir uma estilização do terminal mais complexo. Depende de cada caso. Aqui vamos ver desde o mais simples até o completo, dando liberdade para estilizar o terminal da maneira que quiser. 

Para instalar, basta usar o código

```bash
pip install rich
```

Ou utilize o uv, conda etc.

## Sequência de aulas

Segue a seguir a ordem dos conteúdos que será visto no curso. Um conteúdo não necessáriamente depende do anterior. O Rich trabalha com partes isoladas, mas que se complementam. Portanto pode usar qualquer um dos módulos em conjunto. 

Independente disso, é bom ver o Uso rápido e o Console de qualquer forma e depois seguir a ordem ou ir no conteúdo que mais te interessa. 

Lembrando que utilizei a documentação oficial: (https://rich.readthedocs.io/en/latest/introduction.html) como base, mas esse curso não é um copia e cola da documentação. Explico com mais detalher cada parte e tem exercícios.

### 1 - Uso rápido

Alguns comandos simples para uso rápido. Para quem está querendo um terminal mais legal, mas sem complicar muito o código.

### 2 - Console

O principal módulo do Rich. Ele é a base de tudo, mesmo que alguns comando funcionem sem que defina um objeto da classe Console. Deve-se inicar por aqui, caso queira aprender a biblioteca de verdade. 

### 3 - Text

Para trabalhar com textos longos. Aqui tem controle da estilização de cada parte do texto assim como configurações. 

### 4 - Styles

A melhor forma de definir um estilo reutilizável no Rich. Possui diversas configurações e controle maior dos estilos. Ótimo se quiser algo mais verboso e bem definido, onde o usuário não tem controle sobre o seu programa.

### 5 Theme

Herda todos as configurações do Style e é útil caso queira algo mais semântico e simples ou que dependa do usuário para troca de estilos.

### 6 - Panel

Uma forma legal de visualizar qualquer elemento do Rich ou do Python dentro de uma caixa personalizada. Aqui você pode definir o estilo do painel e do conteúdo interno dela. Alguns métodos do Rich já possum paineis pré-definidos, como traceback e inspect.

### 7 - Table

Visualzação criação de uma tabela com vários estilos diferentes vista dentro do terminal. Possível adcionar e retirar elementos, assim como configurar o esilo de cada um deles.

### 8 - Tree

Visualização melhorada das pastas do seu projeto.

### 9 - Progress

Barra de progresso para uma aplicação. Bem visual e legal.

### Resto

O Rich possui outras finalidade, mas no meu ponto de vista começa a existir outras biblioteca que fazem melhor, como o uso de Layouts para dashbords e o live, para atualização em tempo real, entre outros.
