## style

O par√¢metro mais interessante √© o style. √â um aprimoramento do Console Markup e ser√° utilizado como padr√£o a partir de agora, sozinho ou com o m√≥dulo Style visto posteriormente. 

Com o style, √© poss√≠vel definir o estilo do seu texto ou String de diversas formas poss√≠veis. Veremos v√°rias aqui.

### Cores padr√£o

Ele aceita por padr√£o as cores: black, red, green, yellow, blue, magenta, cyan, white

````python
from rich.console import Console

console = Console()

default_colors: list[str] = ["black", "red", "green", "yellow", "blue", "magenta", "cyan", "white"]

for color in default_colors:
    console.print(f'Oi, [white]com a cor:[/] {color}' ,style=color )
    
````

### Varia√ß√µes bright e dark

As cores padr√£o possuem a varia√ß√£o bright e dark.

dark_cor: ['dark_black', 'dark_red', 'dark_green', 'dark_yellow', 'dark_blue', 'dark_magenta', 'dark_cyan', 'dark_white']

bright_cor: ['dark_red', 'dark_green', 'dark_blue', 'dark_magenta', 'dark_cyan']

````python
from rich.console import Console

console = Console()

bright_colors: list[str] = ['bright_black', 'bright_red', 'bright_green', 'bright_yellow', 'bright_blue', 'bright_magenta', 'bright_cyan', 'bright_white']

for color in bright_colors:
    console.print(f'Oi, [white]com a cor:[/] {color}' ,style=color )
````

````python
from rich.console import Console

console = Console()

dark_colors: list[str] = ['dark_red', 'dark_green', 'dark_blue', 'dark_magenta', 'dark_cyan']

for color in dark_colors:
    console.print(f'Oi, [white]com a cor:[/] {color}' ,style=color )
````

### Estilo de texto 

O console.print tamb√©m aceita esses estilos: bold, dim, italic, underline, blink, reverse, strike

````python
from rich.console import Console

console = Console()

console.print("Teste de cor de fundo", style="underline magenta")
console.print("Teste de cor de fundo", style="italic bold red")
````

### Background

Se quiser colocar uma cor de fundo, basta usar a palavra "on" depois de definiar o estilo e em seguida colocar a cor que quer.

````python
from rich.console import Console

console = Console()

console.print("Teste de cor de fundo", style="magenta on green")
console.print("Teste de cor de fundo", style="black on rgb(22,35,84)")
````

### RGB E HEX

Se n√£o quiser utilizar uma cor padr√£o, basta usar o c√≥digo RGB OU HEX da cor que voc√™ quer. 

Para o RGB precisa colocar: rgb() dentro dos " ".

````python
from rich.console import Console

console = Console()

console.print("Cor RGB", style="rgb(227,144,16)")
console.print("cor_hex", style="#e39410")
````

### Escalas de cinza

O style tam√©m aceita uma escala de cinza que vai de 0 a 100. Basta usar a palavra grey0 a grey100.

````python
from rich.console import Console

console = Console()

for i in range(0, 101, 10):
    console.print(f"[grey{i}]grey{i}[/grey{i}]")
````

## justify

Alinhamento do texto. Aceita os valores: "left", "center", "right" e  "full"

````python
from rich.console import Console

console = Console()

console.print("Texto √† esquerda", justify="left")
console.print("Texto centralizado", justify="center")
console.print("Texto √† direita", justify="right")
console.print("Texto justificado em v√°rias linhas para testar o alinhamento completo.", justify="full")
````

## markup

Ativa/desativa Rich Markup.

````python
from rich.console import Console

console = Console()

console.print("[bold green]OK com markup[/]", markup=True)
console.print("[bold green]OK sem markup[/]", markup=False)
````

## emoji

## overflow

Comportamento ao exceder largura. Aceita os valores "fold", "crop", "ellipsis" e "ignore" 

````python
from rich.console import Console

console = Console()

console.print("Texto muito longo para caber na linha", width=20, overflow="fold")
console.print("Texto muito longo para caber na linha", width=20, overflow="crop")
console.print("Texto muito longo para caber na linha", width=20, overflow="ellipsis")
console.print("Texto muito longo para caber na linha", width=20, overflow="ignore")
````

## no_wrap

Desativa quebra autom√°tica.

````python
from rich.console import Console

console = Console()

console.print(
    "Texto muito longo para caber na linha",
    width=20,
    no_wrap=True
)
````
