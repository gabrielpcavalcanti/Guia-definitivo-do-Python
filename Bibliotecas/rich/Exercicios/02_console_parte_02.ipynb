# Exercícios — Rich Console parte 02

### Exercício 1: Calculadora Colorida (Fácil)

Crie uma calculadora simples que use Console para exibir os resultados com cores diferentes dependendo da operação:

- Soma em verde

- Subtração em amarelo

- Multiplicação em azul

- Divisão em magenta

Use input() para receber os números e a operação, e exiba o resultado formatado.

```python
from rich.console import Console

console = Console()

def sum(num1: float, num2: float) -> float:
    return num1 + num2


def subtraction(num1: float, num2: float) -> float:
    return num1 - num2


def multiplication(num1: float, num2: float) -> float:
    return num1 * num2


def division(num1: float, num2: float) -> float:
    return num1 / num2


def main() -> None:
    res: str = console.input("Qual operação vc quer [yellow](soma, sub, mult ou div): [/]")
    num_01: float = float(console.input("Digite o [bold green]primeiro número:  [/]"))
    num_02: float = float(console.input("Digite o [bold green]segundo número:  [/]"))

    match res:

        case "soma":
            console.print(f'A soma de {num_01} e {num_02} é [cyan]{sum(num_01, num_02)}[/]' )
        
        case "sub":
            console.print(f'A subtração de {num_01} e {num_02} é [cyan]{subtraction(num_01, num_02)}[/]')
        
        case "mult":
            console.print(f'A multiplicação de {num_01} e {num_02} é [cyan]{multiplication(num_01, num_02)}[/]')
        
        case "div":
            console.print(f'A divisão de {num_01} e {num_02} é [cyan]{division(num_01, num_02)}[/]')
        
        case _:
            print("Valor inválido")
 
 
if __name__ == '__main__':
    main()
```
