## 4 - Inspect

Uma forma mais bonita de inspecionar objetos do Python. Fica uma visualizaÃ§Ã£o mais bonita e clara.

Os parÃ¢metros sÃ£o

```Python
rich.inspect(
    obj, 
    *, 
    console=None, 
    title=None, 
    help=False, 
    methods=False, 
    docs=True, 
    private=False,
    under=False, 
    sort=True, 
    all=False, 
    value=True
)
```

````python
from rich.console import Console
from rich import inspect

console = Console()

inspect([])

console.rule(style="black")

inspect([], all=True)
````

````python
from rich.console import Console
from rich import inspect

console = Console()

inspect(42, title="[bold white]int[/]", help=True, methods=True)


````

## 5 - Traceback

Ele permite uma visualizaÃ§Ã£o mais bonita do erro. 

````python
from rich.traceback import install

install()

def divide(a, b):
    return a / b


result = divide(10,0)
````

Pode fazer diversas configuraÃ§Ãµes, como vista abaixo (existem mais), mas o melhor Ã© deixar o padrÃ£o. Ã‰ bom o suficiente para nÃ£o precisar mudar. Mas caso queira, a vontade.

## 6 - Markdown

Casa queira ler um Markdown direto do terminal.

````python
from rich.console import Console
from rich.markdown import Markdown

MARKDOWN = """
# This is an h1
Rich can do a pretty *decent* job of rendering markdown.
1. This is a list item
2. This is another list item
"""

console = Console()

md = Markdown(MARKDOWN)

console.print(md)
````
