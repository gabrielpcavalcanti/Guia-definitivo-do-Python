# Theme

O Theme é outra classe do Rich que permite vc nomear estilos dentro de um dicionário. É uma forma mais simples de utilizar estilos. Não substitui o Style e sim, complementa. 

```python
from rich.console import Console
from rich.theme import Theme

# Definir um tema personalizado
meu_tema = Theme({
    "info": "cyan",
    "warning": "yellow",
    "error": "bold red",
    "success": "bold green",
    "destaque": "bold magenta on black",
    "titulo": "bold blue underline",
    "codigo": "dim cyan"
})

# Criar console com o tema
console = Console(theme=meu_tema)

# Usar os estilos do tema
console.print("Informação importante", style="info")
console.print("Atenção aqui!", style="warning")
console.print("Algo deu errado!", style="error")
console.print("Operação concluída!", style="success")
console.print("Título Principal", style="titulo")
console.print("print('Hello')", style="codigo")
```

Diferente do Style, O Theme aceita o uso dentro do Console Markup

```python
from rich.console import Console
from rich.theme import Theme

tema_app = Theme({
    "erro": "bold red",
    "aviso": "yellow",
    "ok": "green"
})

console = Console(theme=tema_app)

console.print("[erro]Um Anel para a todos governar.[/]")
console.print("[aviso]Um Anel para a todos governar.[/]")
console.print("[ok]Um Anel para a todos governar.[/]")
```

A forma mais poderosa é unir o Style com o Theme. Define os estilos com o Style e salva ele dentro de um tema. 

```python
from rich.console import Console
from rich.theme import Theme
from rich.style import Style

ERRO = Style(
        color="red",
        bgcolor="black",
        bold=True,
        underline=True,
        blink=False
    )

tema_avancado = Theme({
    "erro": ERRO,
    
    "aviso": Style(
        color="yellow",
        bgcolor="blue",
        italic=True
    ),
    "sucesso": Style(
        color="green",
        bold=True,
        dim=False
    ),
    "destaque": Style(
        color="white",
        bgcolor="magenta",
        bold=True,
        underline=True,
        reverse=False
    )
})

console = Console(theme=tema_avancado)

console.print("Erro crítico!", style="erro")
console.print("Atenção aqui", style="aviso")
console.print("Tudo OK!", style="sucesso")
console.print("IMPORTANTE", style="destaque")
```

```python
from rich.style import Style
from rich.theme import Theme
from rich.console import Console

# Criar estilos base que serão reutilizados
estilo_base_erro = Style(color="black", bold=True)
estilo_base_aviso = Style(color="yellow")

# Compor estilos
tema = Theme({
    "erro": estilo_base_erro,
    "erro_sublinhado": estilo_base_erro + Style(underline=True),
    "erro_invertido": estilo_base_erro + Style(reverse=True),
    "aviso": estilo_base_aviso,
    "aviso_dim": estilo_base_aviso + Style(dim=True)
})

console = Console(theme=tema)

console.print("Erro normal", style="erro")
console.print("Erro com sublinhado", style="erro_sublinhado")
console.print("Erro invertido", style="erro_invertido")
console.print("Aviso normal", style="aviso")
console.print("Aviso suave", style="aviso_dim")
```

Para pejetos mais complexos, é bom definir um arquivo so para o Style e outro para os Theme. Assim é fácil o gerenciamento de cada uma das classes. 

Lembrando que para programas em que o usuário tem poder de mudar os estilos, o uso do Theme é mais indicado, assim como uso de arquivos JSON.
