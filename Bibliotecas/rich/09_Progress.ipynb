# Progress

classe Progress da biblioteca Rich permite criar barras de progresso avançadas no terminal. Elas são usadas para acompanhar tarefas demoradas de forma visual, clara e profissional.

Diferente de um simples print() ou contador numérico, o Progress mostra percentual, tempo decorrido, tempo restante, velocidade e pode lidar com múltiplas tarefas ao mesmo tempo.

Usar Progress é melhor do que mensagens manuais porque dá feedback visual imediato ao usuário, melhora a experiência em programas longos, evita prints repetidos no terminal, permite múltiplas tarefas simultâneas e deixa sua aplicação com aparência profissional.

A classe Progress é ideal para:

 - Downloads de arquivos

- Processamento em lote

-  Loops longos

- Análise de dados

- Scripts de automação

- CLIs profissionais


```python
from rich.progress import Progress
import time


with Progress() as progress:
    task = progress.add_task("Processando...", total=100)


for _ in range(100):
    
    time.sleep(0.05)
    progress.update(task, advance=1)
```
