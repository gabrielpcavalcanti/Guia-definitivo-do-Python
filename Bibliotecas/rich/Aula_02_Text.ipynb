# Text e Style

## Text

Text é um objeto de texto estilizado do Rich.

Tudo que o Rich imprime vira um Text internamente. Text resolve o problema de múltiplos estilos no mesmo texto, sem markup. Fica mais profissional,reutilizavel e maleável.

É preciso importar o módulo para poder utilizar assim como qualquer outro módulo do rich.


### 1 - Acrescentar textos com estilos diferentes.

Para isso é preciso criar um objeto do tipo Text e utilizar o método append( ). Assim cada texto pode ter um estilo próprio e sem usar markup.

```python
from rich.console import Console
from rich.text import Text

console = Console()
text = Text()

text.append("Status: ", style="cyan")
text.append("OK", style="bold green")

console.print(text)
```

```python
from rich.console import Console
from rich.text import Text

console = Console()
text = Text()

text.append("ERRO!\n", style="bold red")
text.append("arquivo não encontrado", style="white")

console.print(text)
```

### 2 - Alinhamento e justificação

Utilizando o parâmetro justify, com as opções "left", "center", "right", "full".

```python
from rich.console import Console
from rich.text import Text

console = Console()
text = Text()

texto_exemplo = "No fim da tarde, a praça parecia suspensa no tempo: crianças corriam atrás de pombos, idosos conversavam nos bancos e o sol tingia tudo de laranja. O barulho distante do trânsito se misturava ao som de risadas, criando uma calma improvável no meio da cidade. Era um daqueles momentos simples que, sem aviso, fazem o dia parecer completo."

rr = Text(texto_exemplo, justify="right", style="bold red")
rr.append("\narquivo não encontrado", style="white")

console.print(rr)
```
