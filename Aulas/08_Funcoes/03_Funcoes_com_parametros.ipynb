# Funções com parâmetros

São funções criadas pelo programdor ou built-in que recebem dados para serem processadas dentro da própria função. Esses dados passados funcionam com variáveis dentro da própria função e são chamados de parâmetros. 

A função pode possuir quantos parâmetros quiser. São colocados dentro dos ( ) na hora da criação da função. 

Ao utilizar parâmetros nas funções, é possível ser mais flexivel, evita erros com parâmetros incorretos e fica mais legível o código.

Como sempre é utilizado o Type annotation, aqui não será diferente. Veja a aula 03_Padroes_adotados.


```python
def sum_two_numbers(num1: int, num2: int) -> None:
    print(num1 + num2)


def main() -> None:
    sum_two_numbers(12,2)
 
 
if __name__ == '__main__':
    main()
```

A função sum_two_numbers( ) recebe dois parâmetros do tipo inteiro, num1 e num2. Eles serão utilizados dentro da própria função, no caso, eles serão somados, como o nome da função sugere.

## Parâmetros são posicionais

Os parâmetros podem ser qualquer tipo de dado e devem ser colocados na ordem correta quando foram declarados. Caso a ordem for trocada e os tipos de dados continuaem corretos, a função vai execultar sem nenhum erro, mas com os parâmetros trocados.


```python
def name(nome: str, sobrenome: str) -> None:
    print(f'{nome} {sobrenome}')


def main() -> None:

    name('Rafael','nadal')
    name('Williams','Serena')
 
 
if __name__ == '__main__':
    main()
```

Para evitar esse problema é possível colocar dentro do argumento da função o nome do parâmetro. Com isso, pode colocar na ordem que quiser que o interpretador irá identificar o parâmetro corretamente.

```python
def name(nome: str, sobrenome: str) -> None:
    print(f'{nome} {sobrenome}')


def main() -> None:

    name(nome='Rafael',sobrenome='nadal')
    name(sobrenome='Williams',nome='Serena')
 
 
if __name__ == '__main__':
    main()
```

Observação: Quando estamos criando / declarando uma função, o que fica dentro dos parênteses são os parâmetros. No momento que a função é chamada no programa, o que se tem dentro dos parênteses se chama argumentos. 

## Parâmetros com valores default

As funções podem possuir parâmetros que são padrões, ou seja, eles recebem um valor inicial e não precisam ser colocados no momento da declaração. 

Eles aceitam qualquer tipo de dado. A diferença é que será atribuido um valor defalt para esse parâmetro.

Quando não se declara nada nos argumentos, o valor padrão, caso exista, será utilizado. É possível mudar o valor padrão a qualquer momento na chamada da função. É preciso apenas especificar qual é. Caso não faça isso, a função executa na ordem em que os oargumentos foram colocados. 

```python
def formated_list(list: list[int], end: str='\n') -> None:

    for element in list:
        print(element, end=end)


def main() -> None:
    formated_list([1, 2, 3, 4, 5], end='-')
    formated_list([1, 2, 3, 4, 5])
    formated_list([1, 2, 3, 4, 5], end='*')


if __name__ == '__main__':
    main()
```

Caso a função tenha parâmetros padrão e parâmetros obrigatórios, eles sempre são colocados primeiros na construção da função, depois os opcionais.

```python
def func(var1, var2, var3, var4=3, var=6):
    pass
```
