# Funções com retorno

O retorno na programação em Python é a instrução que encerra a execução de uma função e devolve um valor ao chamador. Ele retornar um valor é devolver uma valor ou um conjunto de valores ao programa.

Para o Python saber que é um retorno, é utilizado dentro da função a palavra reservada `return`. Sempre é bom espcificar qual o retorno da função utilizando o -> com o type annotation. Veja aula 03_Padroes_adotados.

É preciso saber que quando o interpretador encontra um return dentro da função, ela para de ser executada imediatamente, Um valor é enviado de volta para quem chamou a função e o controle volta para o ponto da chamada.

```python
def ex_return(par1, par2) -> str:
    return f'{par1} {par2}'


def main() -> None:
    var_01: str = ex_return("oi,", "tudo bem?")

    print(var_01)
 
 
if __name__ == '__main__':
    main()
```

Observação: funções podem ou não retonar valores, depende da necessidade. No Python, curiosamente, sempre ha um retorno, independente se a função retorna algo ou não. Caso não tenha retorno, ele é considerado do tipo None. 

## Função com vários return

É possível ter vários retuns dentro da mesma função. Quando um deles é encontrado, o programa se encerra. Fuciona como um break dentro de um laço de repetição, por exemplo.

Muito utilizado dentro de if-else-statement. O uso do else perde o sentido porque caso nenhum return seja acionado dentro de uma clausula if ou elif, o último return será acionado, sem a necessidade de se colocar um else. É possível, mas desnecessário.

```python
def grat_than_100(number: int) -> bool:

    if number > 100:
        return True
    return False


def main() -> None:
    
    number: int = int(input("Digite um número: "))

    print(grat_than_100(number))
 
 
if __name__ == '__main__':
    main()
    
```

## Return em vários lugares

Normalmente é preciso fornecer uma variável para que o valor retornado seja armazenado, mas o valor retonado pode ser armazenado dentro de outro parâmetro da função, num loop, condional etc. Isso dá uma liberdade muito interessante ao código.


```python
def sub(num1: int, num2: int) -> int:
    return num1 - num2


def conta(num1: int, num2: int, operacao=sub) -> int:
    return operacao(num1,num2)


def main() -> None:

    print(conta(1,3))
    print(conta(5,6))
    print(conta(9,6))

 
if __name__ == '__main__':
    main()
    
```

## Diferença entre print e return

A função print( ) mostra na tela para o usuário, já o return devolve valor para o código.

Caso tenha que mostrar o valor na tela, é preciso colocar a função que possui o return dentro da função print( )



```python
def sum(a: int, b: int) -> int:
    return a + b


def main() -> None:
    print(sum(10,1))
 

if __name__ == '__main__':
    main()
    
```
