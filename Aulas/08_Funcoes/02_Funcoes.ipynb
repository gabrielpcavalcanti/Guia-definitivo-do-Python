# Funções

A medida que os códigos escritos em qualquer linguagem ficam mais complexos é necessário utilizar uma artimanha para reduzir o tamanho e melhorar a leitura dos códigos. É ai que as funções entram. Elas servem para não ter a necessidade de repitir um certo conjunto de blocos e deixar o código mais limpo e bem escrito.

Uma função são pequenos trechos de código que realizam tarefas específicas.

Elas são muito úteis para executar procedimentos repetidas vezes. 

As funções seguem o princípio DRY - Don`t repete yourself. Você cria um bloco de código que pode ser utilizado em qualquer parte do seu programa ou em até outros lugares, dando origens as bibliotecas. Por esse e outros motivos, as funções são muito importantes.

### Quando uma função é bem escrita

É quando executa uma e somente uma tarefa. Funções que fazem mais de uma coisa precisam ser desmebradas para um melhor funcionamento do código.

Não é porque uma função exuculta uma única tarefa que é precisa ser uma única linha, pelo contrário, uma funcão é um bloco de comando que tem a função de realizar determinado objetivo. Esse bloco de comando pode ter uma outra função, código próprio ou qualquer outra funcionalidade da linguagem.


## Declarando uma função

Funções podem conter características, como parâmetros de entrada e retorno. Mas antes de ver isso, é preciso saber como cria uma função.

Para declarar / criar uma função, usamos a palavra reservada **def** seguida do nome da função mais os ( ), que podem ter parâmetros ou não.

```python
def hello_world() -> None:
    print("Hello World")
```

Nun programa completo (que utilizaremos como padrão), ficaria assim:

```python
def hello_world() -> None:
    print("Hello World")


def main() -> None:
    hello_world()

    
if __name__ == '__main__':
    main()
```

A função hello_world( ) é criada e sua única função é imprimir "Hello World" na tela. Ela é chamada dentro da função main( ) que também é chamada dentro do if-statment.  Perceba que na hora de chamar a função é preciso colocar os ( ). Caso não faça isso, dará erro.

É um exemplo muito simples, mas para ver como funciona o conceito. Quando os exemplos forem ficando mais complexos, fará mais sentido.

Observações:

Suguindo o PEP8 e os padrões adotados, nome de funções são todos em minúsculas e se for palavras compostas separa elas com _;

Depois que o bloco de comando ter terminado, da um espaço de duas linhas para continuar a escrita do código;

Todas as função são criadas antes da função main( ), mesmo que se forem criadas depois não haveria problemas.
